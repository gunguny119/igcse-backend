```python
import firebase_admin
from firebase_admin import credentials
from firebase_admin import storage

cred = credentials.Certificate('igcse-predict-c18eb87031fd.json')
firebase_admin.initialize_app(cred, {'storageBucket': 'igcse-predict.appspot.com'})

bucket = storage.bucket()
```

```python
import os

def check_missing(images, bucket):
    for filename in images:  # screenshots/year/month/component/q1
        if 'component2' in filename:
            real_file = filename + '.png'
            blob = bucket.blob(real_file)
            if not blob.exists():
                print(filename)
                continue
        else:
            i = 1
            while True:
                real_file = f'{filename}_p{i}.png'
                blob = bucket.blob(real_file)
                if not blob.exists():
                    if i == 1:
                        print(filename)
                    break
                i += 1
```

```python
for i in bucket.list_blobs():
    if '.pdf' in i.id or '.DS_Store' in i.id:
        i.delete()
```

```python
for i in [41, 42, 43, 61, 62, 63]:
    check_missing(df[df['component'] == i].ms_path.to_list(), bucket)
```
